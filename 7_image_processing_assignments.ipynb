## Task 1

Write a code to perform Dilation only using libraries like Numpy, PIL


### Algorithm

* Initialize an array with all elements as zero and of size equal to input size. Hint => Use np.zeroes_like()


* Create a padded image. Hint => Use np.zeros()


* Assign the pixels of image to padded image.


* Create 2 for loops to loop through each pixel and convolute.

* If result is greater than 0 output should be 1 else 0




````python
import numpy as np
from PIL import Image
import cv2
import matplotlib.pyplot as plt

def convolve(kernel, bnw):
    #####################Flip kernel suboptimally####################
    kernel_inv = np.copy(kernel)
    kernel_inv[[0,2]] = kernel_inv[[2,0]] 
    kernel_inv[:, [0,2]] = kernel_inv[:, [2,0]] 
    #################################################################
    img = cv2.copyMakeBorder(bnw, 1, 1, 1, 1, cv2.BORDER_CONSTANT)
    res_img = np.copy(bnw)
    for i in range(bnw.shape[0]):
        for j in range(bnw.shape[1]):
            k = img[i:i+kernel.shape[0],j:j+kernel.shape[1]]
            result = (k==kernel)
            final = np.any(result==True)
            if final:
                res_img[i][j] = 1
            else:
                res_img[i][j] = 0
                
            
    return res_img
# Use this kernel
kernel = np.array([[1, 1, 1],
                   [1, 1, 1],
                   [1, 1, 1]])

# Write your code for Dilation here
orig_img = cv2.imread("./images/morph1.jpg", 0) # Read Image
# img = cv2.copyMakeBorder(orig_img, 1, 1, 1, 1, cv2.BORDER_CONSTANT)
orig = cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB)
plt.imshow(orig)
plt.figure()
(tresh,bnw1) = cv2.threshold(orig,127,255,cv2.THRESH_BINARY)
bnw = bnw1/255
plt.imshow(bnw)
plt.figure()
res = convolve(kernel, bnw)
plt.imshow(res)
# print(bnw)


````
